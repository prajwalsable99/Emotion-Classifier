```python
import warnings
# import the opencv library 
import cv2 
from matplotlib import pyplot as plt
import numpy as np
warnings.filterwarnings('ignore')
from tensorflow.keras.models import load_model
# from m
model2=load_model('./emotion-classifier.h5')
def getPrediction(img):
    mappit=  {0: 'angry', 1: 'disgust', 2: 'fear', 3: 'happy', 4: 'neutral', 5: 'sad', 6: 'surprise'}
   
   
    img = cv2.resize(img, (48, 48))
    # img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    
    # print(img)
    # plt.imshow(img,cmap='gray')
    img=img.reshape(-1,48,48,1)
    img=img/255.0
    y_predicted =model2.predict(img)
    y_predicted_class = np.argmax(y_predicted,axis = 1) 
    print(mappit[y_predicted_class[0]])
    return mappit[y_predicted_class[0]]
```

```python
# import the opencv library 
import cv2 
from matplotlib import pyplot as plt

vid = cv2.VideoCapture(0) 
haar_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 


while True:
    ret, frame = vid.read()
    if not ret:
        break
    frame2=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces_rect = haar_cascade.detectMultiScale( frame2, scaleFactor=1.1, minNeighbors=5)
    print(faces_rect)
    font = cv2.FONT_HERSHEY_SIMPLEX
    for (x, y, w, h) in faces_rect: 
        face=frame[y:y+h,x:x+w]
        txt=getPrediction(frame)   
        cv2.putText(face,  txt,  (50, 50),  font, 1,  (0, 255, 255),   2,   cv2.LINE_4) 
    cv2.imshow('frame', frame)
    


    
   
    if (cv2.waitKey(1)==27): 
        break
vid.release()
cv2.destroyAllWindows()
```
